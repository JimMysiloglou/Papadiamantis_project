## Religious

```python
url = 'https://papadiamantis.net/aleksandros-papadiamantis/syggrafiko-ergo/arthra/209-thriskeftika#megali-deftera'

page = requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
divs = soup.find_all("div", class_="accordion-body")

# Extract and print the ids of each div, if it has one
ids = [div.get("id") for div in divs if div.get("id")]
ids
```

```python
chosen_ids = ['i-evdomas-ton-agion-pathon',
 'ta-agia-theofaneia',
 'i-agia-kai-megali-evdomas',
 'i-evdomas-tis-diakainisimou',
 'i-megali-evdomas-en-athinais',
 'kth-iouniou-petros-kai-pavlos-oi-koryfaioi',
 'i-koimisis-tis-theotokou',
 'xristougenna',
 'agiovasileiatika',
 'theofaneia',
 'i-enneakosietiris-tis-pammegistis-lavras',
 'to-makrakeion-epeisodion-en-skiath',
 'teleftaia-apantisis-eis-ton-logon',
 'ta-mnimosyna-kai-to-kathartirion',
 'i-megali-evdomas-en-athinais-2',
 'kyriaki-tou-pasxa',
 'ai-mou-giorgi-epi-t-eort-tis-kg-apriliou',
 'iereis-ton-poleon-kai-iereis-ton-xorion',
 'prologos-kai-simeioseis-eis-to-vivliarion-panagia-i-kounistria-pros-tous-anagnostas',
 'i-dipsa-tou-david',
 'peri-tis-panagias-tis-kounistrias']
```

```python
filtered_divs = [div for div in soup.find_all("div") if div.get("id") in chosen_ids]
```

```python
filtered_divs[0].h2.text
```

```python
print(soup.find('div', id="set-rl_sliders-2").text)
```

```python
religious_articles = dict()
religious_articles[filtered_divs[0].h2.text] = re.sub(r'\s*\(\d{4}\)$', '', soup.find('div', id="set-rl_sliders-2").text.strip())
```

```python
for div in filtered_divs[1:]:
    title = div.h2.text
    content = re.sub(r'\s*\(\d{4}\)$', '', div.text.strip())
    religious_articles[title] = content
```

```python
for title, text in religious_articles.items():
    # Replace spaces or special characters in title to make it filename-safe
    safe_title = "".join(c if c.isalnum() else "_" for c in title)
    
    # Open a text file in write mode
    with open(f"{safe_title}.txt", "w", encoding="utf-8") as file:
        # Write the main text
        file.write(text)
```
